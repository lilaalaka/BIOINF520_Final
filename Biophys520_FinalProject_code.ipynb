```python
import math
import numpy as np
```

```python
# Equation parameters
No = 2       # initial Population Size (units)       
C = 6        # difference between initial and maximum population size
mu = 0.1     # maximum growth rate (units/time)
lmbda = 50   # duration of lag phase (time)
t = 0        # time (initial t=0)

# Equation 1: Modified Gompertz model under static conditions
logN = math.log10(No) + C*math.e**(-math.e**(2.71*mu/C*(lmbda-t)+1))

# Initialize lists for plotting
# Run for loop to generate data points
logN_t = []
t_points = []
for i in range(201):
    t_points.append(i)
    logN = math.log10(No) + C*math.e**(-math.e**(2.71*mu/C*(lmbda-i)+1))
    logN_t.append(logN)

# Plot results
import matplotlib.pyplot as plt
plt.plot(t_points, logN_t,'b-')
plt.xlabel('Time')
plt.ylabel('log(N)')
plt.title('Modified Gompertz Model Under Static Conditions')
plt.show()
```

```python
#Simulating bacterial competition by mixing the Lotka-Volterra predator prey model with the
#logistic growth model 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import integrate
import ipywidgets as ipw

#Defining the Lotka-Volterra model
alpha = 1. 
beta = 1.
delta = 1.
gamma = 1.
x0 = 4.
y0 = 2.

def derivative(X, t, alpha, beta, delta, gamma):
    x, y = X
    dotx = x * (alpha - beta * y)
    doty = y * (-delta + gamma * x)
    return np.array([dotx, doty])

Nt = 1000
tmax = 30.
t = np.linspace(0.,tmax, Nt)
X0 = [x0, y0]
res = integrate.odeint(derivative, X0, t, args = (alpha, beta, delta, gamma))
x, y = res.T

(2, 1000)

plt.figure()
plt.grid()
plt.title("Simulating Bacterial Competitive Evolution")
plt.plot(t, x, 'b', label = 'Bacterial Species 1')
plt.plot(t, y, 'r', label = "Bacterial Species 2")
plt.xlabel('Time')
plt.ylabel('Population Size')
plt.legend()

plt.show()
```

```python
#Creating the model function

import scipy as sci   

def LVC(N, t, r1, r2, alpha, beta, K1, K2):
    """Lotka-Volterra Competition model in Bacteria"""
    dN_1 = r1*N[0] * (1 - (N[0] + alpha * N[1])/ K1)
    dN_2 = r2*N[1] * (1 - (N[1] + beta * N[0])/ K2)
    return(np.array([dN_1, dN_2]))

#3. Set up initial conditions and parameters
r1 = 0.1
r2 = 0.1
alpha = 0.4
beta = 0.66
K1 = 100
K2 = 100

NI = np.array([6, 5]) #Storing N1 and N2 variables

t = np.linspace(0, 150, 1000) #Defining the time variable

sol = sci.integrate.odeint(LVC, NI, t, args=(r1, r2, alpha, beta, K1, K2)) #integrating
t
#Plotting
plt.title("Simulating Bacterial Competition")
plt.plot(t, sol[:, 0], color = "blue", label = "Bacterial Species 1")
plt.plot(t, sol[:, 1], color = "red", label = "Bacterial Species 2")
plt.legend(loc = "upper right")
plt.xlabel("Time")
plt.ylabel("Population Size")
plt.show()
```
